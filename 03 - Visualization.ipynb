```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# two datasets (movies + actor's gender) combined with string variables checked
combined_movies = pd.read_csv('df_movies.csv')
combined_movies = combined_movies.assign(profit = combined_movies.gross - combined_movies.budget)
np.round(combined_movies.sort_values('gross', ascending = True).head(10),2)
```

```python
np.round(combined_movies.describe(),2)
```

```python
# Count of movies in each year by gender
```

```python
np.log(combined_movies['gross']).plot(kind='box')
```

```python
np.log(combined_movies['gross_adj']).plot(kind='hist', bins = 100)
plt.title('Distribution of Gross Revenue')
plt.xlabel('Revenue_adj (log)')

# not log gross revenue distribution: 
#     combined_movies['gross'].plot(kind='hist', bins = 100)
```

```python
np.log(combined_movies['budget_adj']).plot(kind='hist', bins = 100)
plt.title('Distribution of Budget')
plt.xlabel('Budget_adj (log)')
```

```python
# Relationship between the budget and gross revenue 
combined_movies_male = combined_movies[combined_movies["gender"]=="male"]
combined_movies_female = combined_movies[combined_movies["gender"]=="female"]

plt.scatter(combined_movies_male["budget_adj"],y = combined_movies_male['gross_adj'],s=5)
plt.scatter(combined_movies_female["budget_adj"],y = combined_movies_female['gross_adj'], c='tab:red',s=5)
plt.legend(('male', 'female'))
plt.title('Relationship Between Budget and Gross Revenue')
plt.ylabel('Gross Revenue')
plt.xlabel('Budget')
```

```python
# Relationship between the budget and gross revenue (log)

plt.scatter(combined_movies_male["budget_adj"],y = combined_movies_male['gross_adj'], s=5)
plt.scatter(combined_movies_female["budget_adj"],y = combined_movies_female['gross_adj'], c='tab:red',s=5)
plt.legend(('male', 'female'))
plt.yscale('log')
plt.xscale('log')
plt.title('Relationship Between Budget and Gross Revenue (log)')
plt.ylabel('Gross Revenue(log)')
plt.xlabel('Budget(log)')
```

```python
# Relationship between the gender and gross revenue
plt.scatter(combined_movies['year'],y = combined_movies['gross'],s=5)
```

```python
# histogram for average of female and male
gender_mean = combined_movies.groupby("gender").agg([np.mean])
gender_mean[['gross','budget']].plot(kind = 'bar', alpha = 0.25)
plt.title('Mean of Gross Revenue and Budget by Gender')
plt.ylabel('US Dollar (in million)')
plt.legend()
```

```python
combined_movies['log_gross_adj'] = np.log(combined_movies['gross_adj'])
combined_movies[['log_gross_adj','gender']].boxplot(column=['log_gross_adj'], by = 'gender')
plt.show()
```

```python
combined_movies['log_budget_adj'] = np.log(combined_movies['budget_adj'])
combined_movies[['log_budget_adj','gender']].boxplot(column=['log_budget_adj'], by = 'gender')
plt.show()
```

```python
year_count_male = combined_movies.loc[combined_movies['gender']=='male',['year','gender']].groupby(['year']).count()
year_count_female = combined_movies.loc[combined_movies['gender']=='female',['year','gender']].groupby(['year']).count()
year_proportion = year_count_female/(year_count_male + year_count_female)
plt.bar([i for i in range(1980,2021)] , year_proportion.iloc[:,0].values, label = 'Female', alpha = 0.85)
plt.bar([i for i in range(1980,2021)] , (1-year_proportion).iloc[:,0].values, bottom=year_proportion.iloc[:,0].values, label = 'Male', alpha = 0.85)
plt.legend(['Female','Male'])
plt.ylabel('Percentage of Total Movies Within the Year')
plt.show()
```
