```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_inflation = pd.read_csv('inflation and interest.csv')[['country','year','Inflation, consumer prices (annual %)','Inflation, GDP deflator (annual %)']].dropna()

#sort the value in reverse order
df_inflation = df_inflation.iloc[::-1]

#select US market
df_inflation = df_inflation[df_inflation['country'] == 'United States']

# set 2020 as standard year and calculate inflation rate for each country in each year
df_inflation = df_inflation[(df_inflation['year'] >= 1980) & (df_inflation['year'] <= 2020)]

df_inflation['Inflation, consumer prices (annual %)'] = (df_inflation['Inflation, consumer prices (annual %)'] + 100)/100
df_inflation['Inflation, GDP deflator (annual %)'] = (df_inflation['Inflation, GDP deflator (annual %)'] + 100)/100

df_inflation['Inflation, consumer prices cumprod'] = df_inflation.iloc[:,[0,2]].groupby('country').cumprod()
df_inflation['Inflation, GDP deflator cumprod'] = df_inflation.iloc[:,[0,3]].groupby('country').cumprod()

# # set 2020 as standard year
# df_inflation.loc[df_inflation['year']==2020, 'Inflation, GDP deflator cumprod'] = 1
# df_inflation.loc[df_inflation['year']==2020, 'Inflation, consumer prices cumprod'] = 1

df_inflation#.iloc[:,[0,1,4,5]]
```

```python
df_combined = pd.read_csv('df_combined.csv',index_col = 0)
```

```python
df_movies = df_combined.merge(df_inflation.iloc[:,[0,1,4,5]], how = 'inner', on = ['year'])
df_movies
```

```python
# Adding if male
df_movies['if male'] = (df_movies['gender']=='male')

# Adding gross and budget adjusted by inflation
df_movies['gross_adj'] = df_movies['gross']*df_movies['Inflation, GDP deflator cumprod']
df_movies['budget_adj'] = df_movies['budget']*df_movies['Inflation, GDP deflator cumprod']

# Adding profitability
df_movies['profitability'] = df_movies['gross']/df_movies['budget'] - 1
df_movies['profitability_ln'] = np.log(df_movies['gross']/df_movies['budget'])

# drop useless column
df_movies.drop(columns = ['count_movie_names','count_name','country_y'])

# change the year
df_movies['year'] -= 1979

# change scientific 
# df_movies = df_movies[df_movies['gross_adj'] >= 10000000]
df_movies['gross_adj'] = df_movies['gross_adj']/1000000
df_movies['budget_adj'] = df_movies['budget_adj']/1000000
df_movies['gross'] = df_movies['gross']/1000000
df_movies['budget'] = df_movies['budget']/1000000
df_movies = df_movies[df_movies['gross_adj'] >= 10] # This is 10 millions

# Genre adjusting: grouping all other genres except for action, comedy and drama to the Others genre
df_movies['genre_adj'] = df_movies['genre'].copy()
df_movies['genre_adj'].replace([ 'Animation','Biography','Crime','Family','Fantasy','Horror','Mystery','Romance','Sci-Fi','Thriller','Western'],
['Others','Others','Others','Others','Others','Others','Others','Others','Others','Others','Others'], inplace=True)
df_movies.groupby('genre_adj').size()

# Rating adjusting: Grouping the rating G and PG; Grouping the rating R and NC-17 and all the other rating not listed; 
df_movies.groupby('rating').size()
df_movies['rating_adj'] = df_movies['rating'].copy()
df_movies['rating_adj'].replace(['G', 'PG','R','NC-17 and others'], ['G and PG','G and PG','R, NC-17 and others','R, NC-17 and others'], inplace=True)
df_movies.groupby('rating_adj').size()

df_movies.head(3)
```
